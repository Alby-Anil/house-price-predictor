```python
data_encoded = pd.get_dummies(data, dtype=int)

x = data_encoded.drop('median_house_value', axis=1)
y = data_encoded['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train_encoded = pd.get_dummies(x_train, dtype=int)
x_test_encoded = pd.get_dummies(x_test, dtype=int)

# Make sure x_test has all the columns x_train has
x_test_aligned = x_test_encoded.reindex(columns=x_train_encoded.columns, fill_value=0)
```

```python
forest.fit(x_train_encoded, y_train)
forest.score(x_test_aligned, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest=RandomForestRegressor()
param_grid ={
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth": [None,4,8]
}

grid_search= GridSearchCV(forest, param_grid, cv=5,
                          scoring="neg_mean_squared_error",
                          return_train_score=True)

grid_search.fit(x_train,y_train)
```

```python
# Step 1: Use the best model from grid search
best_model = grid_search.best_estimator_

# Step 2: Make predictions on the test set
y_pred = best_model.predict(x_test)

# Step 3: Evaluate model performance
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"RÂ² Score: {r2:.4f}")
```
